```python
# working on genres,cast,crew and keywords, to covert them in to a lost format
```

```python
def convert(obj):
    l=[] #making a list to store
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
con_mov_df["genres"]=con_mov_df["genres"].apply(convert)
```

```python
con_mov_df["genres"]
```

```python
def actor(obj):
    l=[] #making a list to store
    counter= 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter += 1
        else:
            break
    return l
```

```python
# Apply the function to the 'cast' column
con_mov_df["cast"] = con_mov_df["cast"].apply(actor)
```

```python
con_mov_df["keywords"]=con_mov_df["keywords"].apply(convert)
```

```python
con_mov_df.iloc[0].crew
```

```python
def director(obj):
    l=[] #making a list to store
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i["name"])
            break
    return l
```

```python
con_mov_df["crew"]=con_mov_df["crew"].apply(director)
```

```python
con_mov_df.head(50)
```

```python
## Now converting overview from string to list so that we can concatenate it with other lists
con_mov_df["overview"]= con_mov_df["overview"].apply(lambda x:x.split())
```

```python
con_mov_df
```

```python
# Now clear the blank spaces between the words, for further conflict.
con_mov_df['genres'] = con_mov_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
con_mov_df['keywords'] = con_mov_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
con_mov_df['cast'] = con_mov_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
con_mov_df['crew'] = con_mov_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

#### Creating the Tag column by merging up the overview, cast, crew, keywords, and genres column

```python
con_mov_df['tags']= con_mov_df["overview"] + con_mov_df["cast"] + con_mov_df["crew"] + con_mov_df["keywords"] + con_mov_df["genres"]
```

#### Once we create the tag column, then the other 5 column are unnessery for our pourpus, so it's better to delete

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()  # Correctly instantiate the PorterStemmer
```

```python
new_movie= con_mov_df[["id","title","tags"]]
```

```python
new_movie
```

```python
# Now convert the tags column from list to string
new_movie["tags"] = new_movie["tags"].apply(lambda x:" ".join(x))
```

```python
new_movie["tags"].iloc[0]
```

```python
new_movie["tags"] = new_movie["tags"].apply(lambda x:x.lower())
```

```python
def stem(text):
    z = []
    for i in text.split():  # Iterate over each word in the text
        z.append(ps.stem(i))  # Stem the word and append to the list
    return " ".join(z)  # Join the stemmed words into a single string and return
```

```python
new_movie["tags"] = new_movie["tags"].apply(stem)
```

## Vectorisation of the columns

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features = 5000, stop_words ="english")
```

```python
vectors= cv.fit_transform(new_movie["tags"]).toarray()
```

```python
vectors.shape
```

```python
name= cv.get_feature_names_out()
print(name)
```

###### Now I will calculate the cosine distance instead of the Euclidean distance because, in higher-dimensional data, Euclidean distance is not always the best measure. Using cosine distance, I'll determine the angle (θ=theta) between two movies; a smaller angle indicates greater similarity.  

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_bv = cosine_similarity(vectors)
print(type(similarity_bv))
# Change into int32. 
similarity = similarity_bv. astype('int16')  
print(similarity.shape)
print(similarity)
```

```python
def recommendation(movie):
    movie_index = new_movie[new_movie['title']== movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), key=lambda x: x[1], reverse=True)[1:6]


    for i in movie_list:
        print (new_movie.iloc[i[0]].title)
```

```python
recommendation("Toy Story")
```
